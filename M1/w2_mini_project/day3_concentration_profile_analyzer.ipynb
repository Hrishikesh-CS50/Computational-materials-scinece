# Concentration profile analyzer

```python
import numpy as np

x = np.array([0,1,2,3,4], dtype=float)
C = np.array([12,10,8,6,4], dtype=float)

dx = 1.0
D = 0.25

grad = (C[1:] - C[:-1]) / dx
J = -D * grad

print("Gradient:", grad)
print("Flux:", J)
```
