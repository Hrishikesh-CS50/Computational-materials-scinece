# Build Crystal Analyzer


```python
import numpy as np
```

```python
atoms = np.array([
[0,0,0],
[0.5,0.5,0],
[0.5,0,0.5],
[0,0.5,0.5]
])
print("Number of atoms:", len(atoms))

for i in range(len(atoms)):
    print("Atom", i, atoms[i])

d = np.linalg.norm(atoms[1]-atoms[0])
print("Distance atom0-atom1 =", d)
```
