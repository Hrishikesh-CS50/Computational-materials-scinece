# 1D Diffusion-Like Transport Simulation

Today you build your first true computational materials workflow.

```python
# initalize
import numpy as np
import matplotlib.pyplot as plt

C = np.array([10.,10.,10.,0.,0.,0.]) #create array with 6 elements

history = [C.copy()] # create list to store history of C, copy() is used to create a copy of the array
```

```python
# Time Evolution
dt = 0.1 # time step

for step in range(50): # loop over time steps

    C_new = C.copy() # create a copy of C to store new values

    for i in range(1,len(C)-1): # loop over spatial points, skip the first and last point (boundary conditions)

        C_new[i] = C[i] + dt*(C[i+1]-2*C[i]+C[i-1]) # update C_new[i] using the diffusion-like equation

    C = C_new # update C with new values

    history.append(C.copy()) # append a copy of the new C to history
```

```python
# Plot
for state in history[::10]:
    plt.plot(state)

plt.xlabel("Position")
plt.ylabel("Concentration")
plt.title("Diffusion-like Smoothing")
plt.show()
```
