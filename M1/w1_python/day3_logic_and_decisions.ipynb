Is material brittle or ductile?<br>

```python
# Topic 1: Boolean Values

# A boolean has only two states
True
False
```

Examples:

temperature > 300 → True<br>

# Topic 2: Comparison Operators
| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | equal            |
| `!=`     | not equal        |
| `>`      | greater than     |
| `<`      | less than        |
| `>=`     | greater or equal |
| `<=`     | less or equal    |


```python
x = 5
print(x > 3)
print(x == 5)
print(x != 2)
```

# Topic 3: if Statement

```python
temperature = 350

if temperature > 300:
    print("High temperature region") #Indentation is mandatory.
```

# Topic 4: if / else

```python
stress = 150

if stress > 200:
    print("Failure risk")
else:
    print("Safe region")
```

# Topic 5: if / elif / else

```python
T = 900

if T < 273:
    print("Below freezing")
elif T < 600:
    print("Moderate temperature")
else:
    print("High temperature")
```

# Topic 6: Logical Operators
| Operator | Meaning   |
| -------- | --------- |
| `and`    | both true |
| `or`     | one true  |
| `not`    | reverse   |


```python
T = 400
P = 10

if T > 300 and P > 5:
    print("Operating zone")
```
