# Python Loops

```python
# Why Loops Matter

# Instead of writing:
print(1)
print(2)
print(3)
print(4)

# We automate repetition.
```

```python
# Topic 1: for Loop
for i in range(5):
    print(i)
```

```python
range(5)
# 5 is excluded
```

```python
# Topic 2: Controlled Range
for i in range (1,5):
    print(i)
```

```python
# Topic 3: Step Size
for i in range(0,11,2):
    print(i)
```

```python
# Topic 4: Using Loop Variables
for t in range(5):
    temperature = 300 + 10*t
    print(temperature)

# This mimics heating over time steps.
```

```python
# Topic 5: while Loop

# Use when repetition depends on condition

x = 0

# while x < 5:
#     print(x) # No update → never ends.. this is infinite cell, stop the running cell. 
while x < 5:
    x = x + 1
    print(x)
```
