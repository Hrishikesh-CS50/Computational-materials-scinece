Topic 1: User Input

A program can ask the user for data. This is called user input. In Python, we can use the `input()` function to get input from the user. The `input()` function takes a string as an argument, which is the prompt that will be displayed to the user. The function then waits for the user to enter some data and press Enter. The data entered by the user is returned as a string.

```python
name = input("Enter material name: ")
print(name)
```

## Important Rule ##

input() always gives text (string).<br>
Even if user types: 25 , python stores "25". That matters for calculations

## Topic 2: Numbers from Input ##

To calculate, convert text into numbers.

```python
temp = float(input("Enter temperature: "))
print(temp)
```

Why float()?<br>
Use float() when decimals are possible.<br>
Examples:<br>
density = 2.7<br>
voltage = 12.5<br>
pressure = 101.3<br>
int() for Whole Numbers

```python
atoms = int(input("Enter atoms count: "))
print(atoms)
```
