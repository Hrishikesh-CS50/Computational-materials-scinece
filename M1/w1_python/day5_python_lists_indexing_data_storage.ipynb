# Python Lists, Indexing, and Data Storage


Why Lists Matter

A single variable stores one value:

```python
temperature = 300
```

But real science needs many values:

many atoms<br>
many temperatures<br>
many time steps<br>

Use a list.

```python
# Topic 1: Creating Lists
temps = [300, 320, 340, 360]
print(temps)
```

```python
materials = ["Copper", "Steel", "Silicon"]
```

```python
# Topic 2: Indexing

# Python starts counting at 0.

temps = [300,320,340]

print(temps[0])   # 300
print(temps[1])   # 320
print(temps[2])   # 340
```

```python
# Topic 3: Negative Indexing
print(temps[-1])
```

```python
# Topic 4: Slicing
temps = [300,320,340,360,380]

print(temps[1:4])
```

```python
# Topic 5: Append Data

temps = []  
temps.append(300)
temps.append(325)
temps.append(350)
print(temps)

# Used heavily in simulations.
```

```python
# Topic 6: Modify Values
temps[1] = 330
```

```python
# Topic 7: Loop Through Lists
for T in temps:
    print(T)
```

```python
# Topic 8: Length
print(len(temps))
```
