# Material scinece: mass balance


```python
import numpy as np

x = np.array([0,1,2,3], dtype=float)
y = x

dx = 1.0

area = np.sum(y)*dx

print(area)
```

```python
import numpy as np

C = 0.0
dt = 1.0
inflow = 2.0
outflow = 1.0

for t in range(5):
    C = C + dt*(inflow - outflow)
    print("time", t, "C =", C)
```
