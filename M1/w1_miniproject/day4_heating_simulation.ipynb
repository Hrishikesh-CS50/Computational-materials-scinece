# Basic Heating simulation

```python
# Create a simple furnace ramp.

T = 300

for step in range(10):
    print("Step:", step, "Temperature:", T)
    T = T + 25
```

```python
# Stop when temperature exceeds 500.
T = 300
step = 0

while T <= 500:
    print("step:", step, "Temperature:", T)
    T = T + 25
    step = step + 1
print("Final step:", step, "Final Temperature:", T, "which exceeds 500.")
```

```python
# Simulate cooling using a while loop

T = 1000

while T >= 300:
    print(T)
    T = 0.95 * T

print("Final temperature below 300:", T)
```
