# Temperature History Recorder

```python
temps = []

for step in range(6):
    T = 300 + 25*step
    temps.append(T)

print("All temperatures:", temps)
print("First:", temps[0])
print("Last:", temps[-1])

avg = sum(temps)/len(temps)
print(f'Average temperature: {avg}')
```

# Bonus Challenge (University Level)

```python
import numpy as np

# Store atom x-positions
x_positions = np.array([0.5, 1.2, 2.8, 4.0])

# Add 0.1 to each position
new_x_positions = x_positions + 0.1

# Compute displacement
displacement = new_x_positions - x_positions

print("Original positions:", x_positions)
print("Updated positions:", new_x_positions)
print("Displacement:", displacement)
```
