# Materials Science: Coordinates & Atomic Positions
Atoms Need Coordinates

To simulate crystal structures:

Each atom needs position:

Atom	x	y	z
1	0	0	0
2	0.5	0.5	0
3	0.5	0	0.5
Example: Cubic Cell Corner Atoms

Positions:

(0,0,0),(1,0,0),(0,1,0),(0,0,1)

These define geometry.

Why Coordinates Matter

Coordinates determine:

neighbor atoms
bond lengths
crystal symmetry
density
deformation
