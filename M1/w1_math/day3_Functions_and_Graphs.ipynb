# Topic 1: What is a function 

A function maps input to output.

y=f(x)

Example:<br>
y=x**2<br>
Input x gives output y.

```python
# Cell 1: Calculation

import numpy as np

# Create x values
x = np.linspace(-10, 10, 400)

# Define the function
y = x**2
```

```python
# Cell 2: Graph

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x²', linewidth=2)

# Axes lines
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Labels and title
plt.title("Visualization of y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.show()
```

Table of Values<br>
| x  | y |
| -- | - |
| -2 | 4 |
| -1 | 1 |
| 0  | 0 |
| 1  | 1 |
| 2  | 4 |

# Topic 3: Quadratic Functions <br>
y=ax**2+bx+c

Used in:

Energy wells<br>
Optimization<br>
Kinematics<br>

```python
# Cell 1: Calculation
# Set coefficients
a = 1
b = 2
c = 1

# Create x values
x = np.linspace(-6, 4, 400)

# Define quadratic function
y = a*x**2 + b*x + c

# Vertex
h = -b / (2*a)
k = a*h**2 + b*h + c

# Discriminant
D = b**2 - 4*a*c

# Roots
if D > 0:
    r1 = (-b + np.sqrt(D)) / (2*a)
    r2 = (-b - np.sqrt(D)) / (2*a)
elif D == 0:
    r1 = r2 = -b / (2*a)
else:
    r1 = r2 = None
```

```python
# Cell 2: Graph

import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))

# Plot parabola
plt.plot(x, y, label='y = ax² + bx + c', linewidth=2)

# Axes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Vertex
plt.scatter(h, k, s=80, label='Vertex')
plt.text(h+0.2, k+0.5, f'({h:.1f}, {k:.1f})')

# Axis of symmetry
plt.axvline(h, linestyle='--', linewidth=1, label='Axis of Symmetry')

# Roots
if r1 is not None:
    plt.scatter(r1, 0, s=70, label='Root')
    plt.text(r1+0.2, 0.5, f'{r1:.1f}')
    
if r2 is not None and r2 != r1:
    plt.scatter(r2, 0, s=70)
    plt.text(r2+0.2, 0.5, f'{r2:.1f}')

# Labels
plt.title("Quadratic Function with Vertex and Roots")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.show()
```

# Topic 5: Python Graphing Intro

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100) # Create 100 points between -5 and 5
y = x**2 # Define the function y = x^2

plt.plot(x,y)     # Plot the function
plt.xlabel("x")    # Label for x-axis
plt.ylabel("y")    # Label for y-axis
plt.title("y = x^2")  # Title for the plot
plt.show()         # Display the plot
```
