# Vectors

Why Vectors Matter

Scalars = magnitude only

Examples:

temperature<br>
density<br>
energy<br>

Vectors = magnitude + direction

Examples:

force<br>
velocity<br>
displacement<br>
electric field

```python
v = [3,4]
```

```python
# Magnitude of the vector (6,8)
import math
v = [6,8]
magnitude = math.sqrt(v[0]**2 + v[1]**2) # v[0] is the x component, v[1] is the y component
print(magnitude)
```

```python
# Add vectors (1,1) and (2,3)
v1 = [1,1]
v2 = [2,3]
v3 = [v1[0] + v2[0], v1[1] + v2[1]]
print(v3)
```
