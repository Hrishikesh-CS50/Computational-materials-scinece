## Mathematics: Linear Equations

Topic 1: Solve Simple Equations
x+4=10

```python
# Topic 1
print("Topic 1: Solve Simple Equations")
print("x + 4 = 10")
x = 10 - 4
print("x =", x)
print()
```

```python
# Advanced 
from sympy import symbols, Eq, solve

x = symbols('x')

eq = Eq(x + 4, 10)
answer = solve(eq, x)

print(answer)
```

```python
# Topic 2: 3x = 12
eq2 = Eq(3 * x, 12)
answer2 = solve(eq2, x)
print("3x = 12")
print(answer2)
```

```python
# Topic 3: 2x + 5 = 13
eq3 = Eq(2 * x + 5, 13)
answer3 = solve(eq3, x)
print("2x + 5 = 13")
print(answer3)
```
