Topic 3: Algebra

Solve:<br>
x+5=12<br>
Answer: x=7<br>

to solve this with python we import math library called SymPy and usually place the import at the beginning of the notebook

```python
#from sympy import symbols, Eq, solve # place it at the top of the file to avoid redundancy and improve readability and uncomment it

x = symbols('x')             # Define a symbol 'x' for use in equations, this tells python to treat x as a mathematical symbol, not a normal variable with a numeric value.

equation = Eq(x + 5, 12)             # Define an equation: x + 5 = 12: Eq(left_side, right_side) means an equation where the left side is 'x + 5' and the right side is '12'. This sets up the equation we want to solve for 'x'.
answer = solve(equation, x)           # Solve the equation for 'x' and store the answer in the variable 'answer': solve(expression, variable) means:solve this equation for x
print(answer)
```

```python
# original value
old = 100

# new value
new = 120

# percentage increase
increase = ((new - old) / old) * 100

print(increase, "%")
```
