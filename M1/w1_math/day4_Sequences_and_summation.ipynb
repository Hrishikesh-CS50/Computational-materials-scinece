Topic 2: Arithmetic Sequence

Constant difference.

2,4,6,8,10,...



You can use Python to make the arithmetic-sequence formula intuitive by turning it into something you can see, generate, and test.<br>
means:

a1= first term<br>
d = common difference (how much each step changes)<br>
n = which term you want<br>
an​= value of the nth term<br>

```python
a1 = 3
d = 2

for n in range(1, 6):
    an = a1 + (n - 1) * d
    print(f"Term {n}: {an}")
```

```python
# Visual version
a1 = 3
d = 2

for n in range(1, 6):
    an = a1 + (n - 1) * d
    print("■" * an, an)
```
