# Computational solving wiht numpy

```python
# Topic 1: Matrix Solver
import numpy as np

A = np.array([
    [2,1],
    [1,-1]
], dtype=float)

b = np.array([5,1], dtype=float)

x = np.linalg.solve(A,b)

print(x)
```

```python
# This is manual inversion and is said to be less efficient than the above method. So we will not use this method.
import numpy as np

A = np.array([[2,1],[1,-1]], dtype=float)
b = np.array([5,1], dtype=float)

A_inv = np.linalg.inv(A)
x = A_inv @ b

print(x)
```

```python
r = A @ x - b # residual, should be close to zero
print(r)
```

```python
A = np.array([[1,1],[1,-1]])
b = np.array([6,2])

solution = np.linalg.solve(A,b)
print(solution)
```

```python
r = A @ solution - b
print(r)
```
