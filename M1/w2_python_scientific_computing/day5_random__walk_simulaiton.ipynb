# Computation: Random Walk Simulation

```python
# Topic 1: Random Number Generation

import numpy as np
r = np.random.rand()
print(r)
```

```python
# Topic 2: Random Left/Right Motion
```

```python
position = 0

for step in range(10):

    move = np.random.choice([-1, 1])
    position = position + move

    print("Final position:", position)
```

```python
# Topic 3: Store Trajectory

position = 0
trajectory = [position]

for step in range(100):

    move = np.random.choice([-1,1])

    position += move

    trajectory.append(position)

print(trajectory)
```

```python
import matplotlib.pyplot as plt

# -----------------------------
# Random Walk Simulation
# -----------------------------
position = 0
trajectory = [position]

for step in range(100):

    move = np.random.choice([-1, 1])

    position += move

    trajectory.append(position)

# -----------------------------
# Plot Trajectory
# -----------------------------
plt.figure(figsize=(10, 6))

plt.plot(trajectory, linewidth=2)

plt.axhline(0, linestyle='--')

plt.title("1D Random Walk Trajectory")
plt.xlabel("Step Number")
plt.ylabel("Position")

plt.grid(True)

plt.show()
```
