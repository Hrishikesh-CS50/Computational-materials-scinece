# Scientific Computing: Arrays as Physical Data
Why Arrays Matter

Real materials problems involve many values:

atom positions<br>
temperatures<br>
stresses<br>
compositions<br>
grain sizes<br>

Single variables are insufficient.

```python
import numpy as np
```

```python
# 1D arrays
temps = np.array([300, 320, 340, 360])
print(temps)

#This may represent temperature over time.
```

```python
# vectorized operations
temps_c = np.array([0,25,100])
temps_k = temps_c + 273.15
print(temps_k)

# Used for unit conversion.
```

```python
# 2D Arrays
A = np.array([
    [1,2],
    [3,4]
])
print(A)
#Used for matrices, coordinate tables.
```

```python
# Shape

print(A.shape)

#Returns rows and columns.
```
