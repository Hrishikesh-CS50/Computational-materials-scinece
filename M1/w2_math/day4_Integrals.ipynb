```python
#example 1
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 5, 500)
y = x**2

# Choose upper integration limit
a = 3

# Area region
x_fill = np.linspace(0, a, 300)
y_fill = x_fill**2

# Integral function
F = x**3 / 3

# Create figure
fig, ax = plt.subplots(2, 1, figsize=(8, 10))

# -----------------------------------
# TOP: x^2 with shaded integral area
# -----------------------------------
ax[0].plot(x, y, linewidth=2)
ax[0].fill_between(x_fill, y_fill, alpha=0.3)

ax[0].set_title("Function y = x²")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

# Display computed area
area = a**3 / 3
ax[0].text(
    2,
    15,
    f"Integral from 0 to {a} = {area:.2f}",
    fontsize=12
)

# -----------------------------------
# BOTTOM: Integral function
# -----------------------------------
ax[1].plot(x, F, linewidth=2)

ax[1].scatter(a, area, s=80)

ax[1].set_title("Integral Function F(x) = x³ / 3")
ax[1].set_xlabel("x")
ax[1].set_ylabel("F(x)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# Generate x values
# -----------------------------------
x = np.linspace(0, 5, 500)

# Original function
y = 3 * x

# Integral function
F = (3 * x**2) / 2

# Choose integration limit
a = 3

# Area region
x_fill = np.linspace(0, a, 300)
y_fill = 3 * x_fill

# Exact integral value
area = (3 * a**2) / 2

# -----------------------------------
# Create figure
# -----------------------------------
fig, ax = plt.subplots(2, 1, figsize=(8, 10))

# ===================================
# TOP GRAPH
# ===================================
ax[0].plot(x, y, linewidth=2)

# Shade area
ax[0].fill_between(x_fill, y_fill, alpha=0.3)

# Mark point
ax[0].scatter(a, 3*a, s=80)

ax[0].set_title("Function y = 3x")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

ax[0].text(
    1.5,
    10,
    f"Area from 0 to {a} = {area:.2f}",
    fontsize=12
)

ax[0].grid(True)

# ===================================
# BOTTOM GRAPH
# ===================================
ax[1].plot(x, F, linewidth=2)

# Mark corresponding integral value
ax[1].scatter(a, area, s=80)

ax[1].set_title("Integral Function F(x) = (3x²)/2")
ax[1].set_xlabel("x")
ax[1].set_ylabel("F(x)")

ax[1].grid(True)

# -----------------------------------
# Display
# -----------------------------------
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# Function
# -----------------------------------
x = np.linspace(0, 3, 500)
y = x

# Integration limits
a = 0
b = 2

# Region to shade
x_fill = np.linspace(a, b, 300)
y_fill = x_fill

# Exact integral value
area = (b**2)/2 - (a**2)/2

# -----------------------------------
# Plot
# -----------------------------------
plt.figure(figsize=(8,6))

# Plot y = x
plt.plot(x, y, linewidth=2, label='y = x')

# Shade integral area
plt.fill_between(x_fill, y_fill, alpha=0.3)

# Mark boundary points
plt.scatter([a, b], [a, b], s=80)

# Labels and title
plt.title(r'Visualization of $\int_0^2 x\,dx$')
plt.xlabel('x')
plt.ylabel('y')

# Show result
plt.text(
    0.7,
    2.2,
    f'Area = {area:.2f}',
    fontsize=12
)

plt.grid(True)
plt.legend()

plt.show()
```
