# Exercise 1 solutions 


```python
import numpy as np

# Step 1: Coefficient matrix
A = np.array([
    [1, 1],
    [1, -1]
])
```

```python
# Step 2: Constant matrix
b = np.array([
    [6],
    [2]
])
```

```python
print("Matrix A:")
print(A)
```

```python
print("Matrix b:")
print(b)
```

```python
# Step 3: Find determinant of A
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Step 4: Check if the system has a unique solution
if det_A != 0:
    print("The system has a unique solution.")
else:
    print("The system does not have a unique solution.")
```

```python
# Step 4: Find inverse of A
A_inverse = np.linalg.inv(A)
print("Inverse of A:")
print(A_inverse)
```

```python
# Step 5: Multiply inverse of A by b
solution = A_inverse @ b

print("Solution:")
print(solution)
```

# Example 2

```python
# faster method to solve linear equations
import numpy as np

A = np.array([[3,1],[1,1]])
b = np.array([7,3])

solution = np.linalg.solve(A, b)
print(solution)
```
