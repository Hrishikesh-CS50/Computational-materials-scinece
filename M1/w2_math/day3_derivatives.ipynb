# Derivatives

```python
# Exmaple 1 of plotting a function and its derivative using Python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)

y = x**2          # original function
dy_dx = 2*x       # derivative

plt.plot(x, y, label="y = x²")
plt.plot(x, dy_dx, label="dy/dx = 2x") # plotting the derivative

plt.axhline(0)
plt.axvline(0)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = x² and its derivative")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Exmaple 2 of plotting a function and its derivative using Python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)

y = 3*x + 5

# pick a point
x0 = 1
y0 = 3*x0 + 5
slope = 3

# tangent line (same as original line)
tangent = slope * (x - x0) + y0

plt.plot(x, y, label="y = 3x + 5")
plt.plot(x, tangent, '--', label="tangent (same line)")

plt.scatter(x0, y0, label="point")

plt.axhline(0)
plt.axvline(0)

plt.legend()
plt.grid(True)
plt.title("Linear function: tangent equals the function")

plt.show()
```

```python
# Example 3 of plotting a function and its derivative using Python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200)

y = 5*x**3
dy_dx = 15*x**2

plt.plot(x, y, label="y = 5x³")
plt.plot(x, dy_dx, label="dy/dx = 15x²")

plt.axhline(0)
plt.axvline(0)

plt.xlabel("x")
plt.ylabel("y")
plt.title("y = 5x³ and its derivative")

plt.legend()
plt.grid(True)

plt.show()
```
