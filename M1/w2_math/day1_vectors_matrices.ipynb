# Mathematics: Vectors and Matrices

```python
import numpy as np
```

```python
A = np.array([[2,0],[0,3]])
x = np.array([4,5])

print(A @ x)
```

```python
# 1. Magnitude of (6,8) 
v = np.array([6, 8])

magnitude = np.linalg.norm(v)

print(magnitude)

# magnitude = np.sqrt(np.sum(v**2))
```

```python
# 2. Add (2,3)+(1,5)
u = np.array([2,3])
v = np.array([1,5])

w = u + v
print (w)
```

```python
A = np.array([[1,0],
              [0,2]])
x = np.array([3,4])

result = A @ x

print(result)
```
